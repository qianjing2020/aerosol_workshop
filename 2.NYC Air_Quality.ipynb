```python
import pandas as pd 
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/ClarksonAirlab/2019_sensor_workshop/master/Python/Tutorial/Tutorial%20Data/NYC_Air_Data.csv')
```

```python
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Time"] = pd.to_timedelta(df["Time"])

df.head()
```

```python
# describe your data here
df.describe()
```

```python
# describe your single value here
df['PS 19_PM2.5'].describe()
```

```python
df['month'] = pd.DatetimeIndex(df['Date']).month
df['weekday'] = pd.DatetimeIndex(df['Date']).dayofweek
df['day_of_week'] = pd.DatetimeIndex(df['Date']).weekday_name
df['hour']= pd.DatetimeIndex(df['Time']).hour
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
#import matplotlib.patches as mpatches
import matplotlib.ticker as mticker
```

```python
fig, ax1 = plt.subplots()

fig.set_size_inches(11.7, 8.27)


ax1.plot(df['Date'],df['CCNY_PM2.5'], 'k-', lw=3,label='CCNY')
ax1.plot(df['Date'],df['PS 19_PM2.5'], 'g-', lw=2,label='PS 19')
ax1.plot(df['Date'],df['Queens College_PM2.5'], 'm-', lw=2,label='Queens College')

plt.xticks(size = 15)
plt.yticks(size = 15)

plt.title('$PM2.5$ : $in$ $NewYork$ $City$',fontsize=20)
ax1.set_xlabel('$Time$ $Line$',fontsize=16,color='k',labelpad=13)
ax1.set_ylabel('$PM2.5$ : $ug/m^3$',color='k',fontsize=16,labelpad=13)

#ax1.grid(False)

#ax1.legend(loc='best',fontsize=17,markerscale=4)

#for label in ax1.xaxis.get_ticklabels():
#       label.set_rotation(0)
#ax1.xaxis.set_major_locator(mticker.MaxNLocator(7))


#plt.savefig('books_read.png',dpi=300)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)

ax = sns.kdeplot(df['CCNY_PM2.5'], shade=True,legend=True,label='$CCNY$ ')
ax = sns.kdeplot(df['PS 19_PM2.5'], shade=True,legend=True,label='$PS 19$')
ax = sns.kdeplot(df['Queens College_PM2.5'], shade=True,legend=True,label='$Queens College$')

ax.set_xlabel('$PM2.5$ : $ug/m^3$',fontsize=16,labelpad=13)
ax.set_ylabel('Density Functions ',fontsize=16,labelpad=13)
ax.set_title('Probabilty density function of three monitoring stations ',fontsize=16)

ax.set(xlim=(0, None))
plt.xticks(size = 15)
plt.yticks(size = 15)

ax.legend(loc='best',fontsize=17,markerscale=4)
ax.grid(False)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)

sns.boxplot(x="hour", y=df['CCNY_PM2.5'],data=df)

ax.set_xlabel('Hours',color='k',fontsize=16,labelpad=13)
ax.set_ylabel('$PM2.5$ ($ug/m^3$)',color='k',fontsize=16,labelpad=13)

ax.set_title('CCNY Station',fontsize=16)
```

## select your data to pivot 

```python
new=df.groupby([ "hour", "month"]).mean()

new
```

```python
new=new.reset_index()
pt = new.pivot("hour", "month", "CCNY_PM2.5")
```

```python
# A nice matrix table that made for CCNY stations PM2.5 value that took the mean value of each hours in each month
pt
```

```python


f, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(pt, ax=ax,yticklabels=1,cmap="YlOrRd")


ax.set_xlabel('Month of Year ',color='k',fontsize=16,labelpad=13)
ax.set_ylabel('Hour of day  ',color='k',fontsize=16,labelpad=13)

ax.set_title('PM2.5 concentration Heatmap',fontsize=16)
```

```python

g=sns.clustermap(pt, yticklabels=1,cmap="YlOrRd")


ax.set_xlabel('Month of Year ',color='k',fontsize=16,labelpad=13)
ax.set_ylabel('Hour of day  ',color='k',fontsize=16,labelpad=13)

g.fig.suptitle('PM2.5 concentration Heatmap',fontsize=16) 


```
