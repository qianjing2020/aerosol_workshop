```python
import requests

from bs4 import BeautifulSoup

import pandas as pd
```

```python
url_part1='http://www.airqualityontario.com/history/summary.php?start_day='
url_part2='&start_month='
url_part3='&start_year='
url_part4='&my_hour='
url_part5='&Submit=Update&pol=251'

o3=[]
pm25=[]
no2=[]

for i in range( 2018, 2019):
    for j in range( 7, 13):
        for k in range( 1, 32):
            for l in range( 0, 24):
                url=url_part1+str(k)+url_part2+str(j)+url_part3+str(i)+url_part4+str(l)+url_part5
                response = requests.get(url)
                soup = BeautifulSoup(response.text,'html.parser')
                cornwallo3=soup.find("td", {"headers": "p-122 s-56051"})
                cornwallpm25=soup.find("td", {"headers": "p-124 s-56051"})
                cornwallno2=soup.find("td", {"headers": "p-36 s-56051"})
                o3.append(cornwallo3.text)
                pm25.append(cornwallpm25.text)
                no2.append(cornwallno2.text)
  
```

```python
year=[]
month=[]
date=[]
hour=[]
for i in range( 2018, 2019):
    for j in range( 7, 13):
        for k in range( 1, 32):
            for l in range( 0, 24):
                a=str(i)
                b=str(j)
                c=str(k)
                d=str(l)
                year.append(a)
                month.append(b)
                date.append(c)
                hour.append(d)
```

```python
df = pd.DataFrame(year, columns = ['year']) 
df['month']=month
df['date']=date
df['hour']=hour
df['o3']=o3
df['pm2.5']=pm25
df['no2']=no2

df=df.dropna()
df
```
